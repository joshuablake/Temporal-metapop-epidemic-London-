Build a dict, `substitutions`, which maps names to change to what they become and a `frozenset`, `all_locs`, which contains the name of all stations we need to determine the local authority of.

# Wiki data

Add the Wikipedia data to the above dictionary. Expectation is that there is at least one station for each row in the Wikipedia table (sometimes multiple). In practice there are a few which don't match due to normalisation issues, this is resolved using the manual entries.

```python
locs = {i.lower(): i for i in all_locs}
for _, row in wiki_locs[wiki_locs['Station'].notna()].iterrows():
    name = row['Station']
    name = name.replace("'", '')
    name = name.replace('.', '')
    name = name.replace('-', ' ')
    name = name.lower()
    if name in locs:
        station_borough[locs[name]] = row['Local authority']
        continue
    matches = [loc[1] for loc in locs.items() if loc[0].startswith(name)] \
                + [loc[1] for loc in locs.items() if name.startswith(loc[0])]
    if len(matches)== 0:
        print('{} matches nothing'.format(name))
    else:
        for match in matches:
            station_borough[match] = row['Local authority']
        
remaining_locs = all_locs - frozenset(station_borough.keys())
```

# Bus data use

Map as many stations as possible to the bus stop data.

```python
multiple_matches = {}
unfound = []
TRIM = (' nr', ' tram', ' dlr', ' sr')
SUBS = (
    (' rd', ' road'),
    (' brdg', ' bridge'),
    (' jcn', ' junction'),
)
bus_locs['Stop_Name'] = bus_locs['Stop_Name'].str.lower()
for loc in remaining_locs:
    been_found = False
    normalised_loc = loc.lower()
    for s in TRIM:
        if normalised_loc.endswith(s):
            normalised_loc = normalised_loc[:-len(s)]
            
    for start, end in SUBS:
        normalised_loc = normalised_loc.replace(start, end)
    
    matching_bus_stops = bus_locs[bus_locs['Stop_Name'].str.startswith(normalised_loc)]
    matching_bus_stops = matching_bus_stops[matching_bus_stops['Borough'].notna()]
    if matching_bus_stops.empty:
        unfound.append('{} ({})'.format(normalised_loc, loc))
    else:
        matched_boroughs = matching_bus_stops['Borough'].unique()
        if len(matched_boroughs) != 1:
            multiple_matches[loc] = [
                '{} (bus stop: {})'.format(i[1]['Borough'], i[1]['Stop_Name'])
                for i in matching_bus_stops.iterrows()
            ]
        else:
            station_borough[loc] = matched_boroughs[0]
```

```python
for loc, stops in multiple_matches.items():
    print(loc)
    print('\t' + '\n\t'.join(stops))
```

Check that the final stations are the same as the initial stations.

```python
initial_locs = all_locs
final_locs = frozenset(station_borough.keys())

added_locs = final_locs - initial_locs
if added_locs:
    print('Have added: {}'.format(', '.join(added_locs)))
missing_locs = initial_locs - final_locs
if missing_locs:
    print('Have missed: {}'.format(', '.join(missing_locs)))

if not (added_locs or missing_locs):
    print('Looks good!')
```

# Results
